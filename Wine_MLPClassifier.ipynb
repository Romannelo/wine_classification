Загружаем библиотеки

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Загружаем датасет wine

```python
wine = load_wine()
```

Разбиваем на тренировочную и тестовую выборки через встроенные метод селекции (from sklearn.model_selection import train_test_split)

```python
x_train, x_test, y_train, y_test = train_test_split(
    wine.data, 
    wine.target, 
    test_size=0.3, 
    random_state=42
)
```

Инициализируем метод классификации, с параметрами hidden_layer_sizes - размеры скрытых слоев, max_iter - максимальное количество итераций, random_state - для воспроизводимости результатв, verbose=True - выводить процесс обучения в консоль

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(50, 30),  
    max_iter=300,  
    random_state=1,  
    verbose=True,  
)
```

Обучаем модель на тренировочной выборке

```python
mlp.fit(x_train, y_train)
```

Теперь построим график кривой потерь 

```python
plt.figure(figsize=(8, 6))
plt.plot(mlp.loss_curve_, marker='o', linestyle='-', color='b')
plt.title("Кривая потерь (Loss Curve)")
plt.xlabel("Количество итераций")
plt.ylabel("Значение функции потерь")
plt.grid(True)
plt.show()
```

Предскажем классы на тестовой выборке

```python
predictions = mlp.predict(x_test)
```

Оценим точность модели

```python
accuracy = accuracy_score(y_test, predictions)
accuracy
```
